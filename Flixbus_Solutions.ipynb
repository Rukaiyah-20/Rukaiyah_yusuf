```python
import pandas as pd
import numpy as np
```

```python
#read the csv files into python
cust_data = pd.read_csv('customer_challenge.csv')
address_data = pd.read_csv('address_challenge.csv')
country_data = pd.read_csv('country_challenge.csv')
cr_data = pd.read_csv('cr_challenge.csv')
trip_data = pd.read_csv('trip_challenge.csv')
```

```python
# join the required columns
cust_data = cust_data[["CUSTOMER_ID", "CUSTOMER_TYPE", "CUSTOMER_LANGUAGE"]]
address_data = address_data[["ADDRESS_ID", "ZIP_CODE"]]

cr_and_cust = pd.merge(cr_data, cust_data, left_on="CUSTOMER_ID", right_on="CUSTOMER_ID", how="left")
trip_cr_cust = pd.merge(trip_data, cr_and_cust, left_on="CR_ID", right_on="CR_ID", how="left")

trip_cr_cust_address = pd.merge(trip_cr_cust, address_data, left_on="DEPARTURE_ADDRESS_ID", right_on="ADDRESS_ID", how="left")
trip_cr_cust_address = trip_cr_cust_address.rename(columns={'ZIP_CODE': 'DEPARTURE_ZIP_CODE'})

trip_cr_cust_address.drop('ADDRESS_ID', axis=1, inplace=True)
                                                            
trip_cr_cust_address = pd.merge(trip_cr_cust_address, address_data, left_on="ARRIVAL_ADDRESS_ID", right_on="ADDRESS_ID", how="left")
trip_cr_cust_address = trip_cr_cust_address.rename(columns={'ZIP_CODE': 'ARRIVAL_ZIP_CODE'})                                                       

trip_cr_cust_address.drop('ADDRESS_ID', axis=1, inplace=True)
trip_cr_cust_address.drop('DEPARTURE_ADDRESS_ID', axis=1, inplace=True)
trip_cr_cust_address.drop('ARRIVAL_ADDRESS_ID', axis=1, inplace=True)
```

```python
# substring the required part
trip_cr_cust_address["DEPARTURE_UTC"] = trip_cr_cust_address["DEPARTURE_UTC"].astype("str").str[0:10]
trip_cr_cust_address["ARRIVAL_UTC"] = trip_cr_cust_address["ARRIVAL_UTC"].astype("str").str[0:10]
trip_cr_cust_address['DEPT_CROSSING'] = np.where(trip_cr_cust_address['DEPARTURE_ZIP_CODE'].str[0:2] == trip_cr_cust_address['ARRIVAL_ZIP_CODE'].str[0:2], "No", "Yes")
```

```python
trip_cr_cust_address_1 = trip_cr_cust_address[(trip_cr_cust_address["DEPARTURE_UTC"]=="2021-07-31")]
trip_cr_cust_address_2 = trip_cr_cust_address[(trip_cr_cust_address["DEPARTURE_UTC"]=="2021-08-21")]
trip_cr_cust_address_3 = trip_cr_cust_address[(trip_cr_cust_address["ARRIVAL_UTC"]=="2021-07-31")]
trip_cr_cust_address_4 = trip_cr_cust_address[(trip_cr_cust_address["ARRIVAL_UTC"]=="2021-08-21")]

trip_cr_cust_address = trip_cr_cust_address_1.append(trip_cr_cust_address_2)
trip_cr_cust_address = trip_cr_cust_address.append(trip_cr_cust_address_3)
trip_cr_cust_address = trip_cr_cust_address.append(trip_cr_cust_address_4)

trip_cr_cust_address_1 = trip_cr_cust_address[(trip_cr_cust_address["STATUS_VALUE"]=="RESERVED")]
trip_cr_cust_address_2 = trip_cr_cust_address[(trip_cr_cust_address["STATUS_VALUE"]=="IN_PROGRESS")]
trip_cr_cust_address_3 = trip_cr_cust_address[(trip_cr_cust_address["STATUS_VALUE"]=="BOOKED")]

trip_cr_cust_address = trip_cr_cust_address_1.append(trip_cr_cust_address_2)
trip_cr_cust_address = trip_cr_cust_address.append(trip_cr_cust_address_3)
```
